```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

##### Plotting functions

```python
def plotgraph(x,y):
    #making the spines for cartesian plane)
    plt.plot(x,y, color = 'blue')
    plt.gca().spines['bottom'].set_linewidth(2)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(2)
    plt.gca().spines['left'].set_position(('data', 0))
    plt.title()
    plt.xticks(np.arange(-5, 6, 1))
    plt.yticks(np.arange(-5, 6, 1))
    plt.grid(True)
    plt.show()
```

```python
def plotsingraph(a, k, p, q):
    #setting size of box
    #y-range
    pos_y = a+q+1
    neg_y = pos_y*(-1)
    #x-range
    pos_x = 360
    neg_x = -90
    
    #creating function y=sin(x) and y=a sin k(x-p) + q
    x = np.arange(-90, 361, 1)  
    y = np.sin(np.radians(x))
    x1 = np.arange(-90, 361, 1)  
    y2 = a*(np.sin(np.radians(k*(x-p)))) + q
    
    # Plotting the graphs
    # Plotting y=sin(x)
    plt.plot(x,y, color = 'blue')
    plt.plot(x1, y2, color = 'orange')
  
    # Making the spines for cartesian plane)
    plt.gca().spines['bottom'].set_linewidth(2)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(2)
    plt.gca().spines['left'].set_position(('data', 0))
    plt.title('Plot of y=sin(x) and y = a sin k(x-p) + q')
    plt.xticks(np.arange(neg_x, pos_x+1, 45))  # Set grid every 45 degrees
    plt.yticks(np.arange(neg_y+0.5, pos_y+0.5*q, 0.5))
    plt.grid(True)
```

```python
def plotcosgraph(a, k, p, q):
    #setting size of box
    #y-range
    pos_y = a+q+1
    neg_y = pos_y*(-1)
    #x-range
    pos_x = 360
    neg_x = -90
    
    #creating function y=sin(x) and y=a sin k(x-p) + q
    x = np.arange(-90, 361, 1)  
    y = np.cos(np.radians(x))
    x1 = np.arange(-90, 361, 1)  
    y2 = a*(np.cos(np.radians(k*(x-p)))) + q
    
    # Plotting the graphs
    # Plotting y=sin(x)
    plt.plot(x,y, color = 'red')
    plt.plot(x1, y2, color = 'green')
  
    # Making the spines for cartesian plane)
    plt.gca().spines['bottom'].set_linewidth(2)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(2)
    plt.gca().spines['left'].set_position(('data', 0))
    plt.title('Plot of y=cos(x) and y = a cos k(x-p) + q')
    plt.xticks(np.arange(neg_x, pos_x+1, 45))  # Set grid every 45 degrees
    plt.yticks(np.arange(neg_y+0.5, pos_y+0.5*q, 0.5))
    plt.grid(True)
```

```python
def plottangraph(a,k,p,q):
    # x and y values for tan
    x1 = np.arange(2,80, 1)
    x2 = np.arange(100, 260, 1)
    x3 = np.arange(280, 360, 1)
    y2 = np.tan(np.radians(x1))
    y3 = np.tan(np.radians(x2))
    y4 = np.tan(np.radians(x3))
    y5 = a*(np.tan(np.radians(k*(x1-p)))) + q
    y6 = a*(np.tan(np.radians(k*(x2-p)))) + q
    y7 = a*(np.tan(np.radians(k*(x3-p)))) + q

    #making the spines for cartesian plane)
    plt.gca().spines['bottom'].set_linewidth(2)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(2)
    plt.gca().spines['left'].set_position(('data', 0))
       
    #plotting tan
    plt.plot(x1,y2, color = 'magenta')
    plt.plot(x2,y3, color = 'magenta')
    plt.plot(x3,y4, color = 'magenta')

    #plotting new tan
    plt.plot(x1,y5, color = 'purple')
    plt.plot(x2,y6, color = 'purple')
    plt.plot(x3,y7, color = 'purple')
    
    plt.title('Plot of y=tan(x) and y = a tan k(x-p) + q')
    plt.xticks(np.arange(-90, 361, 45))  # Set grid every 45 degrees
    plt.yticks(np.arange(-6, 6, 1))
    plt.grid(True)
    plt.show()
```
