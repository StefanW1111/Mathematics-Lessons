```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Plotting Quadratic Functions: Translation and Transformation

#### Plotting Functions

```python
# y = ax^2 + bx + c
def quadplot1(a, b, c, xmin, xmax):
    x = np.arange(xmin-1, xmax+1, 0.1)
    y = x**2
    y2 = a*(x*x) + b*x + c
    plt.plot(x,y, color = 'blue')
    plt.plot(x,y2, color = 'red')
    plt.gca().spines['bottom'].set_linewidth(1.5)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(1.5)
    plt.gca().spines['left'].set_position(('data', 0))
    plt.xticks(np.arange(xmin-1, xmax+1, 1))
    plt.grid(True)
    plt.title(f"Plot of y={a}x^2 + {b}x + {c}")
    plt.show()

```

```python
# y=a(x-p)^2 + q
def quadplot2(a, p, q, xmin, xmax):
    x = np.arange(xmin-1, xmax+1, 0.1)
    y = x**2
    y2 = a*(x-p)**2 + q
    plt.plot(x,y, color = 'blue')
    plt.plot(x,y2, color = 'red')
    plt.gca().spines['bottom'].set_linewidth(1.5)
    plt.gca().spines['bottom'].set_position(('data', 0))
    plt.gca().spines['left'].set_linewidth(1.5)
    plt.gca().spines['left'].set_position(('data', 0))
    plt.xticks(np.arange(xmin-1, xmax+1, 1))
    plt.grid(True)
    plt.title(f"Plot of y={a}*(x-{p})^2 + {q}")
    plt.show()
```

### Using the functions to plot quadratic functions

#### Plotting quadratic formulas in their standard form

##### quadplot1(a, b, c, xmin, xmax) 

```python
quadplot1(2, 0, 0, -5, 5)
```

#### Plotting quadratic formulas in their Turning Point formula

##### quadplot2(a, p, q, xmin, xmax)

```python
quadplot2(2, 0, 0, -5, 5)
```
